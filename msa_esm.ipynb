```python
import json
from pprint import pprint
```

```python
cd_dict = {}
num_t = 10

with open('blast/PFOS/cd_dict.json', 'r') as f:
    cd_dict = json.load(f, parse_float=float, parse_int=int)

pprint(cd_dict)
```

```python
candidates = []

for key, value in cd_dict.items():
    best = sorted(list(map(float, value.keys())), reverse=True)[:10]
    for b in best:
        candidates.append(cd_dict[key][str(b)][1])

print(list(dict.fromkeys(candidates)))
```
