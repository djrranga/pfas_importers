```python
import requests
from pprint import pprint

# PubChem Compound ID for PFOS: 74483
cid = 74483
url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{cid}/assaysummary/JSON"

res = requests.get(url)
data = res.json()

pprint(data.get('Table', {}).get("Row"))

# Print out assay descriptions and targets
for assay in data.get('Table', {}).get('Rows', []):
    print("Assay Name:", assay.get('AID'))
    print("Description:", assay.get('Description'))
    print("Target", assay.get('TargetName'))
    print("Outcome", assay.get('Outcome'))
    print("--------------")
```
